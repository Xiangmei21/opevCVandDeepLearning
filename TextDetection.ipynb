```python
# Detecting characters
hImg, vImg, x = img.shape
boxes=pytesseract.image_to_boxes(img)
boxes
```

```python
for b in boxes.splitlines():
    print(b)
    b=b.split(' ')
    print(b)
    x, y, w, h = int(b[1]),int(b[2]),int(b[3]),int(b[4])
    cv2.rectangle(img, (x,hImg-y+1),(w,hImg-h),(0,0,255),3)
cv2.imshow('Result',img)
cv2.waitKey(0)
```

```python
#Detecting characters
for b in boxes.splitlines():
    print(b)
    b=b.split(' ')
    print(b)
    x, y, w, h = int(b[1]),int(b[2]),int(b[3]),int(b[4])
    cv2.rectangle(img, (x,hImg-y+1),(w,hImg-h),(0,0,255),3)
    cv2.putText(img, b[0],(x,hImg-y+25),cv2.FONT_HERSHEY_COMPLEX,1,(45,50,240),2)
cv2.imshow('Result',img)
cv2.waitKey(0)
```

```python
#Detecting words
img=cv2.imread('txtimg1.png')
img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
boxes=pytesseract.image_to_data(img)
for x,b in enumerate(boxes.splitlines()):
    if(x!=0):
        b=b.split()
        print(b)
        if(len(b))==12:
            x, y, w, h = int(b[6]),int(b[7]),int(b[8]),int(b[9])
            cv2.rectangle(img, (x,y),(w+x,y+h),(0,0,255),3)
            cv2.putText(img, b[11],(x,y),cv2.FONT_HERSHEY_COMPLEX,1,(45,50,240),2)
cv2.imshow('Result1',img)
cv2.waitKey(0)            
```
